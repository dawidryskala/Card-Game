```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace') 
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
    'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card:
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return f"{self.suit} of {self.rank}"
    
    def set_ace_value(self, new_value):
        self.value = new_value
        
    def get_name_of_card(self):
        return f"{self.suit} of {self.rank}"
        
```

```python
class Deck:
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
    
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def take_one_card(self):
        return self.all_cards.pop(0)
    
        
```

```python
def ask_for_ace_value(name):
    while True:
        
        try:
            ace_value = int(input(f"{name}, choose the value of the Ace, 1 or 11: "))
            if ace_value == 1 or ace_value == 11:
                print(f"{name}, thank you for your choice\n")
                return ace_value
                break
            else:
                print(f"{name}, you need to provide 1 or 11")
                continue
        except:
            print(f"{name} you need to provide 1 or 11")
            continue
```

```python
class Player:
    
    def __init__(self, name):
        
        self.all_card = []
        self.name = name
        self.cards_value = 0
    
    def add_card(self, card):
        
        if card.rank == "Ace":
            ace_value = ask_for_ace_value(self.name)
            card.set_ace_value(ace_value)
            self.all_card.append(card)
        else:
            self.all_card.append(card)
    
    def remove__all_cards(self):
        
        self.all_card = []
        
    def get_value_of_all_cards(self):
        result = 0
        for card in self.all_card:
            result += card.value
        #print(f"Player {self.name} has {result} points")
        return result
    
    def __str__(self):
        result = 0
        for card in self.all_card:
            result += card.value
        #print(f"Player {self.name} has {result} points")
        return f"{self.name} has {result} points"
```

```python
def player_move(name):
    while True:
        
        try:
            player_input = int(input(f"{name}, choose to 1 to take card or 2 to pass: "))
            if player_input == 1 or player_input == 2:
                print(f"{name}, thank you for your choice\n")
                return player_input
                break
            else:
                print(f"{name}, you need to provide 1 or 2")
                continue
        except:
            print(f"{name} you need to provide 1 or 2")
            continue
    
```

```python
def if_continue_game(name):
    while True:
        
        try:
            player_input = int(input(f"{name}, choose to 1 to continue game or 2 to end: "))
            if player_input == 1 or player_input == 2:
                print(f"{name}, thank you for your choice\n")
                
                if player_input == 1:
                    return True
                else:
                    return False
                break
                
            else:
                print(f"{name}, you need to provide 1 or 2")
                continue
        except:
            print(f"{name} you need to provide 1 or 2")
            continue
    
```

```python
game_on = True

player_name = input("Enter your name: ")
player = Player(player_name)
player_cash = 150
    

while game_on and player_cash > 0:
    
    deck_of_cards = Deck()
    deck_of_cards.shuffle()
    player.remove__all_cards()
    
    
    
    dealer = Player("Dealer")
    
    
    for i in range(2):
        dealer.add_card(deck_of_cards.take_one_card())
        
    
    for i in range(2):
        player.add_card(deck_of_cards.take_one_card())
        
    
    print("{:<25} {:>20}".format("Dealer first card:",dealer.all_card[-1].get_name_of_card()))
    print("{:<25} {:>20}".format("Dealer visible points:",dealer.all_card[-1].value))
    
    print(f"\n{player.name} your cards: ")
    for el in player.all_card:
        print("{:<25} {:>13} points".format(el.get_name_of_card(), el.value))
    print("-"*46)
    print("{:<25} {:>13} points".format("Total:",player.get_value_of_all_cards()))
    

    
    player_input = player_move(player.name)
    
    if player_input == 1:
        dealer.add_card(deck_of_cards.take_one_card())
        player.add_card(deck_of_cards.take_one_card())
    else:
        dealer.add_card(deck_of_cards.take_one_card())
        
    print(f"\n{dealer.name}'s cards: ")
    for el in dealer.all_card:
        print("{:<25} {:>13} points".format(el.get_name_of_card(), el.value))
    print("-"*46)
    print("{:<25} {:>13} points".format("Total:",dealer.get_value_of_all_cards()))
    
    print("")
    
    print(f"\n{player.name} your cards: ")
    for el in player.all_card:
        print("{:<25} {:>13} points".format(el.get_name_of_card(), el.value))
    print("-"*46)
    print("{:<25} {:>13} points".format("Total:",player.get_value_of_all_cards()))
    
    if dealer.get_value_of_all_cards() > 21 and player.get_value_of_all_cards() <= 21:
        print("\nYou won")
        player_cash += 100
        print(f"You have {player_cash} $")
        game_on = if_continue_game(player.name)
        continue
        
    elif dealer.get_value_of_all_cards() <= 21 and player.get_value_of_all_cards() > 21:
        print("\nYou lost")
        player_cash -= 100
        print(f"You have {player_cash} $")
        game_on = if_continue_game(player.name)
        continue
        
    elif dealer.get_value_of_all_cards() <= 21 and  player.get_value_of_all_cards() <= 21:
        
        if dealer.get_value_of_all_cards() == player.get_value_of_all_cards():
            print("\nDraw")
            player_cash -= 100
            print(f"You have {player_cash} $")
            game_on = if_continue_game(player.name)
            continue
            
        elif dealer.get_value_of_all_cards() > player.get_value_of_all_cards():
            print("\nYou lost")
            player_cash -= 100
            print(f"You have {player_cash} $")
            game_on = if_continue_game(player.name)
            continue
        
        elif dealer.get_value_of_all_cards() < player.get_value_of_all_cards():
            print("\nYou won")
            player_cash += 100
            print(f"You have {player_cash} $")
            game_on = if_continue_game(player.name)
            continue       
        
print("\n")
print("-"*46)
print("End of game")
print(f"Your score is {player_cash} $")

          
```
